```python
import os
def test_img(filename):
  %matplotlib inline
  results = model("/content/drive/MyDrive/Colab Notebooks/fruitdetection_yolo/yolov5/test images/"+filename)
  results.print()
  plt.imshow(np.squeeze(results.render()))
```
